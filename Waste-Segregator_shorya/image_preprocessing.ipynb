```python
"""
Program to resize the images in order to save computing power on my laptop during
training and testing and also to convert the data from coco to voc format.
"""

import matplotlib.pyplot as plt
import numpy as np
from PIL import Image
import os, sys
import json
from collections import defaultdict
```

```python
anns_file_path = 'data/annotations.json'

# Read annotations
f=open(anns_file_path, 'r',encoding="utf8")
dataset = json.loads(f.read())
f.close()

categories = dataset['categories']
anns = dataset['annotations']
imgs = dataset['images']

#intializing the images and their data
images={}
images_bbox = defaultdict(list)
images_labels= defaultdict(list)

for i in imgs:
    images[i['id']]=[i['file_name'],i['width'],i['height']]
    
for j in anns:
    cocobox = j['bbox']
    voc_bbox=[cocobox[0],cocobox[1],cocobox[0]+cocobox[2],cocobox[1]+cocobox[3],j['id']]
    images_bbox[j['image_id']].append(voc_bbox)
    images_labels[j['image_id']].append(categories[j['category_id']]['supercategory'])
print(len(anns))
#resizing the data and updating it
size = (256,224)
for key,value in images.items():
    url='data/'+value[0]
    im=Image.open(url)
    box=images_bbox[key]
    labels=images_labels[key]
    im=im.resize(size)
    #im.save('resized_'+url)
    for j in box:
        j[0]=(j[0]*size[0])/value[1]
        j[2]=(j[2]*size[0])/value[1]
        j[1]=(j[1]*size[1])/value[2]
        j[3]=(j[3]*size[1])/value[2]
        print(j)
        if(j[4]<369):
            dataset['annotations'][j[4]-1]['bbox']=[j[0],j[1],j[2],j[3]]
        if(j[4]>371 and j[4]<857):
            dataset['annotations'][j[4]-4]['bbox']=[j[0],j[1],j[2],j[3]]
        if(j[4]>859 and j[4]<2181):
            dataset['annotations'][j[4]-7]['bbox']=[j[0],j[1],j[2],j[3]]
        if(j[4]>2186):
           dataset['annotations'][j[4]-13]['bbox']=[j[0],j[1],j[2],j[3]]
        
with open("annotations.json", "wt") as jsonFile:
    json.dump(str(dataset), jsonFile)


'''
Due to the null values in the data, dumping the data in json creates errors,so i made it into a string and dumped it 
into the annotations file.I then used Word to replace all the single quotations marks with double quotation marks and 
all None values to null.I also had to convert the file into utf-8 encoding whenever i load becuase word chnaged its encoding.
'''
```
